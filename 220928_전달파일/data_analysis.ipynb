```python
import csv
```

```python
f = open('seoul_data.csv', 'r', encoding='utf-8')
data = csv.reader(f)
print(data)
         
```

```python
header = next(data)
# print(header)
```

```python
min = next(data)
min_temp = min[-2]
min_temp = float(min_temp)
print(type(min_temp))

# 추운날 데이터를 리스트에 모우는 곳
minData = []

for row in data:
    row[-2] = float(row[-2])
    minData.append(row[-2])
    
    # 최소값 비교
    if min_temp > row[-2]:
        min_temp = row[-2]
        # 현재 날짜
        date = row[-5]
    
print('기상관측 이래 제일 추운날은', date, '기온은', min_temp)
print(minData)
```
