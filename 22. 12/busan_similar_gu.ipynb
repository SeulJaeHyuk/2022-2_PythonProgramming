```python
import csv
import numpy as np
f = open('busan_age.csv', encoding='utf-8')
data = csv.reader(f)
# print(data)
next(data)
header = next(data)
print(header)


data = list(data)
# print(data)
min = 1
result = []
result_name = ''

home = [] # 입력 리스트
name = input('부산의 구를 입력하세요 : ') 
for row in data:
    # print(row[3:104])
    if name in row[0]:
        for i in row[3:104]:
            # home.append(int(i))
            home = np.array(row[3:104], dtype=int) / int(row[2])
# print(home)

for row in data:
    away = np.array(row[3:104], dtype=int) / int(row[2])
    s = np.sum((home - away)**2)
    if s < min and name not in row[0]:
        min = s 
        result_name = row[0]
        result = away
        print(result_name, result)
       

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.rc('font', family='AppleGothic')
plt.title(name + ' 지역의 인구 구조')
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.legend()
plt.show()
    
```
