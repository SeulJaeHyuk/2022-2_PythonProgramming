```python
import csv
f = open('age_mf.csv', encoding='utf8')
data = csv.reader(f)
m = []
f = []

# 입력 받기
name = input('원하는 부산의 구를 입력하세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m_data = int(i.replace(',', ''))
            m.append(m_data)
        for i in row[106:]:
            f_data = int(i.replace(',', ''))
            f.append(-f_data)
        break
            

# for row in data:
#     if '부산진구' in row[0]:
#         for i in range(0, 101):
#             m_data = int(row[i+3].replace(',', ''))
#             f_data = int(row[-(i+1)].replace(',', ''))
#             m.append(m_data)
#             f.append(f_data)
# f.reverse()
print(f)

import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name)
plt.barh(range(101), m, label='남자수')
plt.barh(range(101), f, label='여자수')
plt.legend()
plt.show()
```
