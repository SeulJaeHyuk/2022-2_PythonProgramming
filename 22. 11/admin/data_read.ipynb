```python
import csv
f = open('age.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data)

result = []
name = input('원하는 구를 입력하세요!')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',', '')
            result.append(int(i))
        # break
print(len(result))
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic') # 한글처리 Mac, 'Malgun Gothic' 윈도우즈
plt.rcParams['axes.unicode_minus'] = False # - 부호 나오기
plt.title(name)
# plt.plot(result, 'g^--', label='인구수')

# 막대그래프 그리기 : bar
plt.barh(range(101), result, color='r', label='인구수')  # range(10) : [0,1,2,....,9]

plt.legend()
plt.show()
```
