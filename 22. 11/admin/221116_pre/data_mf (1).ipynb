```python
import csv
f = open('age_mf.csv', encoding='utf8')
data = csv.reader(f)
result = []  # [450, 340]

# 입력 받기
name = input('원하는 부산의 구를 입력하세요 : ')

for row in data:
    m = 0
    f = 0
    if name in row[0]:
        for i in row[3:104]:
            m += int(i.replace(',', ''))
        result.append(m)
        
        for i in row[106:]:
            f += int(i.replace(',', ''))
        result.append(f)
        break
print(result)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic') 
# Win: Malgun Gothic'
plt.title(name)
label = ['남자수', '여자수']
plt.pie(result, labels=label, counterclock=False, autopct='%.1f%%', 
        explode=(0,0.05), startangle=90)
plt.legend()
plt.axis('equal')
plt.show()
```
