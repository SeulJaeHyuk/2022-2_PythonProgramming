```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic') # mac은 AppleGothic / Window는 Malgun Gothic을 사용한다.

size = ([10, 20, 40, 70])
label = ['A형', 'B형', 'C형', 'O형']

plt.pie(size, labels=label, counterclock=False, autopct='%.1f%%', explode=(0, 0, 0.1, 0), startangle=0) # %%가 하나임.
plt.legend()
plt.axis('equal')
plt.show()
```
