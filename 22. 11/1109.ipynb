```python
import csv
f = open('age.csv',encoding='utf-8')  ## age.csv란 파일을 불러오며, 해당 파일을 utf-8의 형태로 인코딩한다.
data = csv.reader(f)   ## data를 csv 파일로 읽는다.

next(data)

부산진구 = []
해운대구 = []
영도구 = []
동래구 = []

for row in data :
    if '부산진구' in row[0] :
        for i in row[3:] :
            부산진구.append(int(i.replace(',','')))
    if '해운대구' in row[0] :
        for i in row[3:] :
            해운대구.append(int(i.replace(',','')))
    if '영도구' in row[0] :
        for i in row[3:] :
            영도구.append(int(i.replace(',','')))
    if '동래구' in row[0] :
        for i in row[3:] :
            동래구.append(int(i.replace(',','')))
            
```

```python
import matplotlib.pyplot as plt
plt.style.use ('ggplot')
plt.rc('font', family = "Malgun Gothic")
plt.plot(부산진구, color = "r", label = '부산진구', marker='o')
plt.plot(해운대구, color = "b", label = '해운대구', marker='x')
plt.plot(동래구, color = "g", label = '동래구', marker='^')
plt.plot(영도구, color = "y", label = '영도구', marker='s')
plt.legend()
plt.show()
```
