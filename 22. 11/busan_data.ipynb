```python
import csv
f = open('busan_age.csv', encoding='utf-8')
data = csv.reader(f)
print(data)
header = next(data)

m = []
f = []

for row in data :
    if '부산진구' in row[0]:
        # print(row)
        for i in range(3, 104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))

import matplotlib.pyplot as plt
plt.scatter(m, f)
plt.show()
```
